```python
from bs4 import BeautifulSoup as bs
from splinter import Browser
import time
import datetime
import pandas as pd
import lxml
```

```python
#gets alt tags for images that are otherwise given in table as NaN, for levels and country
def getAlt(x):
    for img in x.find_all('img'):
        if(img.get("alt")):
#             print(img.get("alt"))
            img.string = img.get("alt")

#copied code from scrape_mars.py
def getLeaderboardData(lowest_rank):
    url = "https://achievementstats.com/index.php?action=rankings"

    executable_path = {'executable_path': 'chromedriver.exe'}
    browser = Browser('chrome', **executable_path, headless=False)

    browser.visit(url)
    
    #add a delay so page fully loads
    time.sleep(6)
    
    
    #start collecting data
    
#     for i in range(lowest_rank//30 + 2): #200000//30+2
    curr_rank = int(browser.find_by_tag('tr').last.find_by_tag("td").first.text)
    while(curr_rank < lowest_rank):
        
        try:
            current_leaderboard = browser.html

            news_soup = bs(current_leaderboard, "html.parser")

            top_scorers = news_soup.find_all("table")[0]
#             print(top_scorers.prettify())
            getAlt(top_scorers)

            #if(i == 0):
            #else:
                #output = output.append(pd.read_html(top_scorers.prettify(), header=0, flavor="bs4")[0], ignore_index=True)
            curr_rank = int(browser.find_by_tag('tr').last.find_by_tag("td").first.text)
            print(curr_rank, lowest_rank)
            browser.find_by_tag('tr').last.click()
        except:
#             print("Missing page #{} of 100".format(i))
            print("Error loading, try again!")
        
        time.sleep(1)
        
    #case where the rank desired is less than 30
    if(lowest_rank <= 30 and curr_rank <= 30):
        current_leaderboard = browser.html

        news_soup = bs(current_leaderboard, "html.parser")
        top_scorers = news_soup.find_all("table")[0]
        getAlt(top_scorers)
        
    time.sleep(10)
    #read output at the end, only once!
    output = pd.read_html(top_scorers.prettify(), header=0, flavor="bs4")[0]
    
    browser.quit()

    return output
```

```python
cols = ["Rank", "Player Id", "Player Name", "Best achievements","Points","#","Games","100%", "All", "Badges", "XP", "Member since","Hours", "Last update"]

#note that it currently takes about a day to get to ~10000
# get the top 30 ranked players, and then put it at the beginning of the data file for the first ~200k ranked players
output_df = getLeaderboardData(30)
output_df.columns = cols
output_df
```
